```python
#Que9. Find the largest of three numbers.
num1 = int(input("Enter First Number:"))
num2 = int(input("Enter Second Number:"))
num3 = int(input("Enter Third Number:"))
if num1 > num2 and num1 > num3:
    print(f'{num1}: is greatest')
elif num2 > num1 and num2 > num3:
    print(f'{num2}: is greatest')
else:
    print(f'{num3}: is greatest')
```

```python
#Que10. Check if a letter is a vowel or consonant.
letter = input("Enter Letter:")
vowel = ["A","E","I","O","U"]
if letter in vowel:
    print(f'{letter}: is vowel')
else:
    print(f'{letter}: is consonant')
```

```python
#Que11. Check if a number is positive, negative, or zero.
num = int(input("Write a number:"))
if num > 0:
    print(f'{num}: is positive')
elif num < 0:
    print(f'{num}: is negative')
else:
    print(f'{num}: is Zero')
```

```python
#Que1. Simple Calculator: Perform +, -, *, / operations based on user input.
num1 = float(input("Enter First Number:"))
operations = input("Enter Operations:")
num2 = float(input("Enter Second Number:"))
w = (num1 + num2)
x = (num1 - num2)
y = (num1 * num2)
z = (num1 / num2)
if operations == "+":
    print(f': {num1} + {num2} = {w}')
elif operations == "-" :
    print(f': {num1} - {num2} = {x}')
elif operations == "*":
    print(f': {num1} * {num2} = {y}')
elif operations == "/":
    print(f': {num1} / {num2} = {z}')
```
