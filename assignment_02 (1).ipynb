```python
#Q9. Take two numbers as input from the user and display their sum, difference, product, and division.

num1 = float(input("Enter first number:"))
num2 = float(input("Enter second number:"))
sum = num1 + num2
diff = num1 - num2
prod = num1 * num2
div = num1 / num2 if num2 != 0 else "Error: Division by zero"
print("Results:")
print(f"Sum: {num1} + {num2} = {sum}")
print(f"Difference: {num1} - {num2} = {diff}")
print(f"Product: {num1} * {num2} = {prod}")
print(f"Division: {num1} / {num2} = {div}")
```

```python
#Q10. Write a program to calculate the average of 5 numbers entered by the user using arithmetic operators.

a = float(input("Enter first number: "))
b = float(input("Enter second number: "))
c = float(input("Enter third number: "))
d = float(input("Enter fourth number: "))
e = float(input("Enter fifth number: "))

sum=a+b+c+d+e

average=sum/5

print("average of number:",average)
```
